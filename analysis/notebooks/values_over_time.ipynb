```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
```

```python
w_df = pd.read_csv("../data/weight-over-time.csv")
w_df
```

```python
w_df.info()
```

```python
w_df['Datetime'] = pd.to_datetime(w_df['Datetime'])
w_df = w_df[w_df['Datetime'] > datetime(year=2022, month=1, day=1)]
```

```python
w_df.info()
```

```python
# plotting weight over time
plt.plot(w_df['Datetime'], w_df['Value'])
plt.xlabel("Year")
plt.ylabel("Weight (lbs)")
plt.title("Weight over Time")
plt.show()
```
